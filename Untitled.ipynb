```python
trace1 = go.Scatter(
        x = df['Date'].dt.strftime(date_format='%Y-%m-%d'),
        y = df['Close'])
layout = go.Layout(dict(title = 'Closing prices of Apple',
                       xaxis = dict(title = 'Date'),
                       yaxis = dict(title = 'Price (USD)'),
                       ))
data = [trace1]
py.iplot(dict(data=data,layout=layout))
```

```python
df['price_diff'] = df['Close'] - df['Open']
```

```python
trace1 = go.Scatter(
        x = df['Date'].dt.strftime(date_format='%Y-%m-%d'),
        y = df['price_diff'])
layout = go.Layout(dict(title = 'Price change within a day',
                       xaxis = dict(title = 'Date'),
                       yaxis = dict(title = 'Price_diff'),
                       ))
data = [trace1]
py.iplot(dict(data=data,layout=layout))
```

```python
grouped = df.groupby('Date').agg({'price_diff': ['min']}).reset_index()
```

```python
g = grouped.sort_values(('price_diff', 'min'), ascending=False)[:10]
g['min_text'] = 'Maximum price drop: ' + (-1 * g['price_diff']['min']).astype(str)
trace = go.Scatter(
    x = g['Date'].dt.strftime(date_format='%Y-%m-%d').values,
    y = g['price_diff']['min'].values,
    mode='markers',
    marker=dict(
        size = g['price_diff']['min'].values,
        color = g['price_diff']['min'].values,
        colorscale='Portland',
        showscale=True
    ),
    text = g['min_text'].values
    #text = f"Maximum price drop: {g['price_diff']['min'].values}"
    #g['time'].dt.strftime(date_format='%Y-%m-%d').values
)
data = [trace]

layout= go.Layout(
    autosize= True,
    title= 'Top 10 months by standard deviation of price change within a day',
    hovermode= 'closest',
    yaxis=dict(
        title= 'price_diff',
        ticklen= 5,
        gridwidth= 2,
    ),
    showlegend= False
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig,filename='scatter2010')
```

Moving Average for Apple

```python
df['Close_10'] = df['Close'].rolling(10).mean()
df['Close_50'] = df['Close'].rolling(50).mean()
```

```python
data = []
for col in ['Close', 'Close_10', 'Close_50']:
    data.append(go.Scatter(
        x = df['Date'].dt.strftime(date_format='%Y-%m-%d').values,
        y = df[col].values,
        name = col
    ))
    
layout = go.Layout(dict(title = "Trend of values",
                  xaxis = dict(title = 'Month'),
                  yaxis = dict(title = 'Price (USD)'),
                  ),legend=dict(
                orientation="h"),)
py.iplot(dict(data=data, layout=layout))
```

Check if the stock prices of tech companies are correlated

```python
msft = pd.read_excel("MSFT Historical Data.xlsx")
nflx = pd.read_excel("NFLX Historical Data.xlsx")
```

```python
tech_df = df['Date'].copy()
tech_df = pd.DataFrame(tech_df)
tech_df['AAPL'] = df['Close']
tech_df['MSFT'] = msft['Price']
tech_df['NFLX'] = nflx['Price']
tech_df
```

```python
data = []
for col in ['AAPL', 'MSFT', 'NFLX']:
    data.append(go.Scatter(
        x = tech_df['Date'].dt.strftime(date_format='%Y-%m-%d').values,
        y = tech_df[col].values,
        name = col
    ))
    
layout = go.Layout(dict(title = "Trend of values",
                  xaxis = dict(title = 'Month'),
                  yaxis = dict(title = 'Price (USD)'),
                  ),legend=dict(
                orientation="h"),)
py.iplot(dict(data=data, layout=layout))
```

```python
Price_change_df = df['Date'].copy()
Price_change_df = pd.DataFrame(Price_change_df)
Price_change_df['AAPL_change'] = df['Change %']
Price_change_df['MSFT_change'] = msft['Change %']
Price_change_df['NFLX_change'] = nflx['Change %']
Price_change_df
```

```python
data = []
for col in ['AAPL_change', 'MSFT_change', 'NFLX_change']:
    data.append(go.Scatter(
        x = Price_change_df['Date'].dt.strftime(date_format='%Y-%m-%d').values,
        y = Price_change_df[col].values,
        name = col
    ))
    
layout = go.Layout(dict(title = "Trend of values",
                  xaxis = dict(title = 'Month'),
                  yaxis = dict(title = 'Price (USD)'),
                  ),legend=dict(
                orientation="h"),)
py.iplot(dict(data=data, layout=layout))
```

```python
#Heatmap Daily Return
plt.figure(figsize = (16,5))
sns.heatmap(Price_change_df.corr(),annot=True,cmap='summer',linewidths=1,linecolor='k',square=True,mask=False, vmin=-1, vmax=1)
```

```python
# Heatmap closing price
plt.figure(figsize = (16,5))
sns.heatmap(tech_df.corr(),annot=True,cmap='summer',linewidths=1,linecolor='k',square=True,mask=False, vmin=-1, vmax=1)
```
